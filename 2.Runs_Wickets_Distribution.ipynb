### Visualise the distribution with change in location and scale parameters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
from scipy.stats import norm
```

```python
# Parameters for the first normal distribution
mean1 = 0 # Location Parameter
std_dev1 = 1 # Shape Parameter

# Parameters for the second normal distribution
mean2 = 2 # Location Parameter
std_dev2 = 1 # Shape Parameter

# Generate range of x values
x = np.linspace(-5,5,1000)

# Calculate the PDF values for both distributions
pdf1 = norm.pdf(x, mean1, std_dev1)
pdf2 = norm.pdf(x, mean2, std_dev2)

# Plot

plt.figure(figsize=(8,5))
# plot the pdfs for both distributions
plt.plot(x, pdf1, label=f'$\mu={mean1}, \sigma={std_dev1}$', color='blue')
plt.plot(x, pdf2, label=f'$\mu={mean2}, \sigma={std_dev2}$', color='green')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend();
```

```python
# Data Generation
# Parameter Estimation
# Visualisation

# Generate data
np.random.seed(40)
mean_sales = 200
std_dev_sales = 30
num_days = 365

daily_sales = np.random.normal(mean_sales, std_dev_sales, num_days)

# Fit the normal distribution to the data

mu,sigma = stats.norm.fit(daily_sales)

# print the estimated parameters
print(f"Estimated mean (mu): {mu}")
print(f"Estimated standard deviation (sigma): {sigma}")

#Plot the histogram of daily sales and fitted normal distribution 

plt.figure(figsize=(8,5))
plt.hist(daily_sales, bins=20, density=True, alpha=0.6, color='g', label='Daily Sales')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x,mu,sigma)
plt.plot(x,p,'k',linewidth=2, label='Fitted Normal Distribution')
plt.title('Histogram of Daily sales with fitted normal distribution')
plt.xlabel('Daily Sales')
plt.ylabel('Density')
plt.legend();
```

```python
import scipy.stats as st
def get_best_distribution(data):
    dist_names
    dist_results = []
    params = {}
    for dist_name in dist_names:
        dist = getattr(st, dist_name)
        param = dist.fit(data)
        params[dist_name] = param
        
        # Applying the Kolmogorv - smirnov test (ks test)
        d,p = st.kstest(data, dist_name, args = param)
        print("p value for " + dist_name + " = "+str(p))
        dist_results.append((dist_name,p))
    
    #Select the best fitted distribution
    best_dist, best_p = (max(dist_results, key=lambda item: item[1]))
    
    print("Best fitted distribution: "+ str(best_dist))
    print("Bets p value: "+ str(best_p))
    print("Parameters for the best fit: "+str(params[best_dist]))
    
    return best_dist, best_p, params[best_dist]
```

## Create a function to predict values based on the fitted distribution

```python
def predict_beta_dist(a,b, loc, scale, x):\
    return stats.beta.cdf(x,a,b,loc,scale)
```
